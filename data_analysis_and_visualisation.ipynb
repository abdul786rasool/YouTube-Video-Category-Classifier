```python
import seaborn as sns
from matplotlib import pyplot as plt
```

# Category Distribution Count Plot

```python
fig = plt.figure(figsize=(8,6))
pal=sns.cubehelix_palette(9,start=0.5, rot=0, dark=0.3, light=0.9, reverse=True)
sns.set_style("darkgrid")
sns.countplot(data=data,x="category",palette=pal,order=data["category"].value_counts().index)
plt.xticks(rotation="vertical")
plt.show
```

```python
data2=data.groupby(by="category")
```

```python
data2.get_group("music")
```

```python
list(data["category"].unique())
```

## ViewCount density plots in all categories

```python
categories = list(data["category"].unique())
fig, ax = plt.subplots(nrows=3, ncols=3,figsize=(14,9))
fig.suptitle("ViewCount density plots in all categories", fontsize=15)
cm = plt.get_cmap('hsv')
for i,cat in enumerate(categories):
  row = i//3
  col = i%3
  d = data2.get_group(cat)["viewCount"] + 1 #to deal with zeroes
  sns.kdeplot(x=d,ax=ax[row,col],color=cm(i*1/8),log_scale=True,fill=True)
  ax[row,col].legend([cat],loc="upper right" if cat=="news and politics" else "upper left")
fig.tight_layout()
plt.show()
```

## LikeCount density plots in all categories

```python
categories = list(data["category"].unique())
fig, ax = plt.subplots(nrows=3, ncols=3,figsize=(14,9))
fig.suptitle("LikeCount density plots in all categories", fontsize=15)
cm = plt.get_cmap('turbo')
for i,cat in enumerate(categories):
  row = i//3
  col = i%3
  d = data2.get_group(cat)["likeCount"] + 1 #to deal with zeroes
  sns.kdeplot(x=d,ax=ax[row,col],color=cm(i*1/8),log_scale=True,fill=True)
  ax[row,col].legend([cat],loc="upper right" if cat in ["education","news and politics","other"] else "upper left")
fig.tight_layout()
plt.show()
```

## CommentCount density plots in all categories

```python
categories = list(data["category"].unique())
fig, ax = plt.subplots(nrows=3, ncols=3,figsize=(14,9))
fig.suptitle("CommentCount density plots in all categories", fontsize=15)
cm = plt.get_cmap('nipy_spectral')
for i,cat in enumerate(categories):
  row = i//3
  col = i%3
  d = data2.get_group(cat)["commentCount"] + 1 #to deal with zeroes
  sns.kdeplot(x=d,ax=ax[row,col],color=cm(i*1/9),log_scale=True,fill=True)
  ax[row,col].legend([cat],loc="upper left" if cat in ["film/animation"] else "upper right")
fig.tight_layout()
plt.show()
```

## Calculating Average View, Like, and Comment Counts for Each Category

```python
import numpy as np
categories = list(data["category"].unique())
avg_viewCount=[]
avg_likeCount=[]
avg_commentCount=[]
for cat in categories:
  d = data2.get_group(cat)
  avg_viewCount.append(np.mean(d["viewCount"]))
  avg_likeCount.append(np.mean(d["likeCount"]))
  avg_commentCount.append(np.mean(d["commentCount"]))
```

## avg_ViewCounts in all categories

```python
fig = plt.figure(figsize=(14,9))
fig.suptitle("avg_ViewCounts in all categories", fontsize=15)
index=np.argsort(avg_viewCount)[::-1]
cm = sns.cubehelix_palette(9,start=1.8, rot=0, dark=0.2, light=0.6, reverse=True,as_cmap=True)
colors = [cm(i*1/8) for i in range(9)]
x=np.array(categories)[index]
y=np.array(avg_viewCount)[index]
plt.bar(x=x,height=y,log=True,color=colors)
for i in range(len(x)):
    plt.text(i, y[i], int(y[i]), ha = 'center')
fig.tight_layout()
plt.show()
```

# avg_LikeCounts in all categories

```python
fig = plt.figure(figsize=(14,9))
fig.suptitle("avg_LikeCounts in all categories", fontsize=15)
index=np.argsort(avg_likeCount)[::-1]
cm = sns.cubehelix_palette(9,start=1, rot=0, dark=0.2, light=0.6, reverse=True,as_cmap=True)
colors = [cm(i*1/8) for i in range(9)]
x=np.array(categories)[index]
y=np.array(avg_likeCount)[index]
plt.bar(x=x,height=y,log=True,color=colors)
for i in range(len(x)):
    plt.text(i, y[i], int(y[i]), ha = 'center')
fig.tight_layout()
plt.show()
```

## avg_CommentCounts in all categories

```python
fig = plt.figure(figsize=(14,9))
fig.suptitle("avg_commentCounts in all categories", fontsize=15)
index=np.argsort(avg_commentCount)[::-1]
cm = sns.cubehelix_palette(9,start=3, rot=0, dark=0.2, light=0.6, reverse=True,as_cmap=True)
colors = [cm(i*1/8) for i in range(9)]
x=np.array(categories)[index]
y=np.array(avg_commentCount)[index]
plt.bar(x=x,height=y,log=True,color=colors)
for i in range(len(x)):
    plt.text(i, y[i], int(y[i]), ha = 'center')
fig.tight_layout()
plt.show()
```
